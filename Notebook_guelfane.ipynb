```python
# The one code :

import pandas as pd
import re
import os

# Liste des mots-clés SQL
SQL_KEYWORDS = {
    "ADD", "ALL", "ALTER", "AND", "ANY", "AS", "ASC", "BACKUP", "BETWEEN", "CASE", "CHECK", "COLUMN",
    "CONSTRAINT", "CREATE", "DATABASE", "DEFAULT", "DELETE", "DESC", "DISTINCT", "DROP", "EXEC", "EXISTS",
    "FOREIGN", "FROM", "FULL", "GROUP", "HAVING", "IN", "INDEX", "INNER", "INSERT", "IS", "JOIN", "LEFT",
    "LIKE", "LIMIT", "NOT", "NULL", "OR", "ORDER", "OUTER", "PRIMARY", "PROCEDURE", "RIGHT", "ROWNUM",
    "SELECT", "SET", "TABLE", "TOP", "TRUNCATE", "UNION", "UNIQUE", "UPDATE", "VALUES", "VIEW", "WHERE"
}

def clean_column_names(df):
    """Nettoie et normalise les noms de colonnes en remplaçant les espaces par des underscores,
    en mettant en minuscules, en ajoutant un suffixe aux mots-clés SQL et en gérant les doublons."""
    cleaned_columns = {}
    seen_columns = set()

    for col in df.columns:
        # Normalisation des noms de colonnes
        new_col = re.sub(r'\s+', '_', col).strip().lower()

        # Éviter les conflits avec les mots-clés SQL
        if new_col.upper() in SQL_KEYWORDS:
            new_col += '_col'

        # Gérer les doublons de noms de colonnes
        original_new_col = new_col
        counter = 1
        while new_col in seen_columns:
            new_col = f"{original_new_col}_{counter}"
            counter += 1

        cleaned_columns[col] = new_col
        seen_columns.add(new_col)

    df.rename(columns=cleaned_columns, inplace=True)
    return df

def remove_similar_columns(df):
    """Supprime les colonnes similaires en utilisant la logique SQL (noms similaires après normalisation)."""
    normalized_columns = {}
    columns_to_drop = []

    for col in df.columns:
        normalized_col = re.sub(r'\s+', '_', col).strip().lower()
        if normalized_col in normalized_columns:
            columns_to_drop.append(col)
        else:
            normalized_columns[normalized_col] = col

    df.drop(columns=columns_to_drop, inplace=True)
    return df

def load_and_clean_csv(file_path, output_path):
    """Charge un fichier CSV, nettoie les noms de colonnes, supprime les colonnes similaires et le sauvegarde dans le répertoire de sortie."""
    df = pd.read_csv(file_path, sep=';')
    print(f"\nFichier : {file_path}")
    print("Avant nettoyage des colonnes :")
    print(df.head())

    df = clean_column_names(df)
    df = remove_similar_columns(df)

    print("\nAprès nettoyage des colonnes et suppression des doublons :")
    print(df.head())

    df.to_csv(output_path, sep=';', index=False)
    return df

# Chemin du répertoire contenant les fichiers CSV
input_directory = '/data/extractions/site/2024/April'  # Répertoire source
output_directory = '/home/promo-ssh/preprocessed_data/site'  # Répertoire de destination

# Créer le répertoire de sortie s'il n'existe pas
os.makedirs(output_directory, exist_ok=True)

# Liste des fichiers CSV dans le répertoire d'entrée
csv_files = [f for f in os.listdir(input_directory) if f.endswith('.csv')]

# Charger et nettoyer chaque fichier CSV
for csv_file in csv_files:
    input_path = os.path.join(input_directory, csv_file)
    output_path = os.path.join(output_directory, csv_file)
    load_and_clean_csv(input_path, output_path)
```

```python
#!/bin/bash

# Fonction pour analyser un fichier CSV
check_csv() {
    file="$1"
    total_lines=$(cat "$file" | wc -l)
    num_columns=$(head -n 1 "$file" | awk -F ',' '{print NF}')

    echo "Fichier: $file"
    echo "Nombre total de lignes: $total_lines"
    echo "Nombre total de colonnes: $num_columns"

    for ((i=1; i<=$num_columns; i++)); do
        empty_cells=$(awk -F ',' -v col="$i" '{if ($col == "") count++} END {print count}' "$file")
        empty_cells_percentage=$(awk -v empty="$empty_cells" -v total="$total_lines" 'BEGIN {print (empty / total) * 100}')
        echo "Colonne $i : Cases vides : $empty_cells ($(printf "%.2f" $empty_cells_percentage)%)"
    done

    echo -e "\nAperçu des lignes vides :"
    awk -F ',' 'NF==0 {print NR}' "$file"
    echo "------------------------------------"
}

# Vérifie si le dossier est fourni
if [ -z "$1" ]; then
    echo "Usage: $0 <dossier>"
    exit 1
fi

# Dossier contenant les fichiers CSV
dossier="$1"

# Vérifie si le dossier existe
if [ ! -d "$dossier" ]; then
    echo "Le dossier spécifié n'existe pas."
    exit 1
fi

# Crée ou vide le fichier de résultat
result_file="resultat.txt"
: > "$result_file"

# Parcours tous les fichiers CSV dans le dossier et analyse chacun
for file in "$dossier"/*.csv; do
    if [ -f "$file" ]; then
        check_csv "$file" | tee -a "$result_file"
    fi
done

echo "Les résultats ont été sauvegardés dans le fichier '$result_file'."
```

```python
def _get_store_and_origin_from_clusters(row, cluster_df, sep=";"):
    store_codes = row['store_codes_extended']
 # case 1
    if store_codes == store_codes:
        # store_codes is not NAN
        store_codes = [
            int(e.strip()) for e in store_codes.split(sep) if e.strip().isnumeric()
        ]
        origins = [DEFAULT_STORE_CODE_ORIGIN for _ in store_codes]
        return store_codes, origins
```

```python


def enrich_mapping_leaflet_df(mapping_leaflet_df, cluster_df):
    final_df = mapping_leaflet_df.copy()
    final_df["STORE_CODE"] = final_df.apply(
        lambda row: _get_store_and_origin_from_clusters(row, cluster_df, ";")[0], axis=1
    )
    final_df["STORE_CODE_ORIGIN"] = final_df.apply(
        lambda row: _get_store_and_origin_from_clusters(row, cluster_df, ";")[1], axis=1
    )
    final_df = final_df.explode(column=["STORE_CODE", "STORE_CODE_ORIGIN"])
    final_df.reset_index(drop=True, inplace=True)
    comments = final_df["COMMENTS"]
    final_df.drop(columns=["COMMENTS"], inplace=True)
    final_df["PROMOTION_PRIORITY"] = final_df["TYPE_OF_PROMOTION"].apply(
        lambda t: PROMOTION_PRIORITY.get(t, 0)
    )
    final_df["COMMENTS"] = comments
    return final_df.sort_values(
        by=["PROMO_YEAR", "FILE_NAME", "SHEET_NAME", "STORE_CODE"]
    )


def _deduplicate_store_promotions(group):
    sorted_group = group.sort_values(by="PROMOTION_PRIORITY", ascending=False)
    return sorted_group.iloc[0]
```

```python


enriched_mapping_leaflet_df = enrich_mapping_leaflet_df(mapping_leaflet_df, cluster_df)

```
